```python
#SNAKES AND LADDERS GAME FOR 2 PLAYERS:
import random
snakes={17:7,54:34,62:19,98:79}
ladders={3:38,24:33,42:93,72:84}

print("###### Welcome to Snakes & Ladders Game ######")
#Enter names of the players
name1=input("Enter Player 1 name ")
name2=input("Enter Player 2 name ")

#As both players players will start from 0 position1 represents player1 and position2 respresents player2
position1=0
position2=0

#Used for determining which player turn is to participate in the game
chance=1
print("Let us Start")
k=0
#Used for determing which mode player wants to play auto or manual
roll1=input(name1+" Which mode auto or manual if auto press 'roll' for manual press 'man' ")
if(roll1!='roll' and roll1!='man'):
    roll1=input("Wrong Choice, Enter Again 'roll' or 'man' ")
    
roll2=input(name2+" Which mode auto or manual if auto press 'roll' for manual press 'man' ")   
if(roll2!='roll' and roll2!='man'):
    roll2=input("Wrong Choice, Enter Again 'roll' or 'man' ")
    
roll='no'
while(position1<100 and position2<100):
    if(k!=0):
        roll=input(name1+" want to quit? enter 'quit' or press any other key ")#if player1 wants to quit in the middle of the game
        
    if(roll=='quit'):
        print(name2,"won the game")
        print("Game Successfully Finished")
        break
        
    if(roll1=='roll'and chance==1):# player1 rule in snakes and ladders game for auto mode
        print(name1+" chance")
        dice=random.randint(1,6)
        if(position1+dice>100):#checking overshoot condition
            position1=position1
        else:#checking for snakes,ladders..
            if(snakes.get(position1+dice,'none')!='none'):
                   position1=snakes.get(position1+dice)
            elif(ladders.get(position1+dice,'none')!='none'):
                position1=ladders.get(position1+dice)
            else:
                position1+=dice
        print(name1,"got",dice)
        print(name1,"Final Position",position1)
        chance=2 #alternating the chance for both players
        
    if(roll1=='man'and chance==1):#player1 rule in snakes and ladders game for manual mode
        print(name1+" chance")
        num=int(input("Enter a number between 1 and 20 "))
        if(num>20 or num<=0):#prompting user for valid input
            print("Wrong choice")
            num=int(input("Enter Again"))
        if(num>=1 and num<=20):#for checking valid number inputs
            if(num+position1>100):#checking overshoot condition
                position1=position1
            else:#checking for snakes,ladders..
                if(snakes.get(position1+num,'none')!='none'):
                    position1=snakes.get(position1+num)
                elif(ladders.get(position1+num,'none')!='none'):
                    position1=ladders.get(position1+num)
                else:
                    position1+=num
        if(num>=1 and num<=20):           
            print(name1,"got:",num)
        print(name1,"Final Position:",position1)
        chance=2
        
    if(position1==100):#Wining criteria for player1
        print(name1,"won the game")
        print("Game Successfully Finished")
        break
    
    if(k!=0):
        roll=input(name2+" want to 'quit'? enter 'quit' or press any other key ")
    if(roll=='quit'):#if player2 wants to quit in the middle of the game
        print(name1,"won the game")
        print("Game Successfully Finished")
        break
        
        
    if(roll2=='roll'and chance==2):#player2 rule in snakes and ladders game for auto mode
        print(name2+" chance")
        dice=random.randint(1,6)
        if(position2+dice>100):#checking overshoot condition
            position2=position2
        else:#checking for snakes,ladders..
            if(snakes.get(position2+dice,'none')!='none'):
                position2=snakes.get(position2+dice)
            elif(ladders.get(position2+dice,'none')!='none'):
                position2=ladders.get(position2+dice)
            else:
                position2+=dice
        print(name2,"got:",dice)
        print(name2,"Final Position:",position2)
        chance=1
        
    if(roll2=='man' and chance==2 ):#player2 rule in snakes and ladders game for manual mode
        print(name2+" chance")
        num=int(input("Enter a number between 1 and 20"))
        if(num>20 or num<=0):#prompting user for valid input
            print("Wrong choice ")
            num=int(input("Enter Again"))
        if(num>=1 and num<=20):#for checking valid inputs    
            if(num+position2>100):#checking overshoot condition
                position2=position2
            else:#checking for snakes,ladders..
                if(snakes.get(position2+num,'none')!='none'):
                    position2=snakes.get(position2+num)
                elif(ladders.get(position2+num,'none')!='none'):
                    position2=ladders.get(position2+num)
                else:
                    position2+=num
        if(num>=1 and num<=20):            
            print(name2,"got:",num)
        print(name2,"Final Position:",position2)
        chance=1
                
                
    if(position2==100):#Wining criteria for player2
        print(name2,"won the game")
        print("Game Succesfully Finished")
        break        
    k=1    
    
```
